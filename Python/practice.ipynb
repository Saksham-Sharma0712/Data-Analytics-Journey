```python
# Write a Python program to check if a variable is of a specific data type.
x = eval(input("Enter a value: "))

if isinstance(x, int):
    print("Integer")
elif isinstance(x, float):
    print("Float")
elif isinstance(x, bool):
    print("Boolean")
elif isinstance(x, str):
    print("String")
else:
    print("Other data type")
```

```python
# WAP to check string is palindrome or not 
s = input("Enter string")
if s == s[::-1]:
    print(f"{s} is palindrome")
else:
    print(f"{s} is NOT a palindrome")
```

```python
#Simple Calc
num1 = float(input("Enter val of num1"))
num2 = float(input("ENter val of num2"))
operation = input("Enter opn + , - , * , / :")
if operation == "+":
    res = num1 + num2
elif operation == "-":
    res = num1 - num2
elif operation == "*":
    res = num1 * num2
elif operation == "/":
    if num2 != 0:
        res = num1/num2
    else:
        print("Num2 is not div by zero")
else:
    print("Invalid opn:")
print(res)
```
